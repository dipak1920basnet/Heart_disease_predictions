```python
import numpy as np 
import pandas as pd 
from sklearn import preprocessing
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\dipak\OneDrive\desktop\AIML\Heart_disease_predictions\framingham (1).csv")
```

```python
data.head()
```

```python
print(data.columns)
```

```python
# Here the column education doesnt contribute any value to the heart disease so it is dropped
```

```python
data.drop(columns=["education"], inplace = True)
```

```python
data.rename(columns = {"male":"Sex_male"}, inplace=True)
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.head()
```

```python
data.TenYearCHD.value_counts()
```

```python
X = np.asarray(data[["age","Sex_male","cigsPerDay","totChol","sysBP","glucose"]])
```

```python
y = np.array(data["TenYearCHD"])
```

```python
print(X[:6])
```

```python
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
print("Train Set: ",X_train.shape, y_train.shape)
print("Test Set: ", X_test.shape, y_test.shape)
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x="TenYearCHD", data=data, palette="BuGn_r")
plt.title("Count Plot")
plt.savefig("charts/Count_plot.svg")
plt.show()
```

The count plot shows that majority of individuals over 3000 don't have heart diseases while around 500 of them have heart diseases. 

```python
latest = data["TenYearCHD"].plot()
plt.savefig("charts/count_num")
plt.show(latest)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model  = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
```

```python
print("The model accuracy score is: ", accuracy)
```
